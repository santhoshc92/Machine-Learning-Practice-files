```python
import pandas as pd
import numpy as np

np.random.seed(1212) #it makes "random" numbers predictable and repeatable.

import keras
from keras.models import Model
from keras.layers import *
from keras import optimizers
from keras.utils import to_categorical
```

```python
df_train = pd.read_csv(r'D:\intellipaat ds ai ml\DS@AI\6 AI\000 - self paced\Module 4 - Modelling with Keras\Dataset\train.csv')
df_test = pd.read_csv(r'D:\intellipaat ds ai ml\DS@AI\6 AI\000 - self paced\Module 4 - Modelling with Keras\Dataset\test.csv')
```

```python
df_train.head() # 784 features, 1 label
```

```python
df_features = df_train.iloc[:, 1:785]
df_label = df_train.iloc[:, 0]

X_test = df_test.iloc[:, 0:784]

print(X_test.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_cv, y_train, y_cv = train_test_split(df_features, df_label, 
                                                test_size = 0.2,
                                                random_state = 1212)

X_train = X_train.to_numpy().reshape(33600, 784) #(33600, 784)    
X_cv = X_cv.to_numpy().reshape(8400, 784) #(8400, 784)

X_test = X_test.to_numpy().reshape(28000, 784)

# X_train = X_train.as_matrix().reshape(33600, 784) #(33600, 784)    as_matrix deprecated. so use to_numpy()
# X_cv = X_cv.as_matrix().reshape(8400, 784) #(8400, 784)

# X_test = X_test.as_matrix().reshape(28000, 784)
```

```python
print((min(X_train[1]), max(X_train[1])))
```

```python
X_train[0]  #1st image in x_train which shows pixel values (0-255) of the 1st image  .same for 2nd- [1], 3rd- [2]
```

```python
# Feature Normalization (pixel values are b/w 0 to 255) 
X_train = X_train.astype('float32'); X_cv= X_cv.astype('float32'); X_test = X_test.astype('float32')
X_train /= 255; X_cv /= 255; X_test /= 255
# the values in each row are pixel intensity values, which typically range from 0 to 255 for grayscale images:

# 0 → black

# 255 → white

# values in between → different shades of gray
```

```python
# Convert labels to One Hot Encoded
num_digits = 10
y_train =to_categorical(y_train, num_digits)
y_cv = to_categorical(y_cv, num_digits)
```

```python
# Printing 2 examples of labels after conversion
print(y_train[0]) # 2
print(y_train[3]) # 7
```

```python
# Input Parameters
n_input = 784 # number of features
n_hidden_1 = 300
n_hidden_2 = 100
n_hidden_3 = 100
n_hidden_4 = 200
num_digits = 10
```

```python
Inp = Input(shape=(784,))
x = Dense(n_hidden_1, activation='relu', name = "Hidden_Layer_1")(Inp) #Inp → Hidden_Layer_1 → x 
x = Dense(n_hidden_2, activation='relu', name = "Hidden_Layer_2")(x) #chain of connections is created by repeatedly calling layers with ()
x = Dense(n_hidden_3, activation='relu', name = "Hidden_Layer_3")(x)
x = Dense(n_hidden_4, activation='relu', name = "Hidden_Layer_4")(x)
output = Dense(num_digits, activation='softmax', name = "Output_Layer")(x)
```

```python
# Our model would have '6' layers - input layer, 4 hidden layer and 1 output layer
model = Model(Inp, output)
# model = Model(inputs =Inp, outputs =output)
model.summary() # We have 297,910 parameters to estimate
```

```python
# Insert Hyperparameters
learning_rate = 0.1
training_epochs = 20
batch_size = 100
sgd = optimizers.SGD(learning_rate=learning_rate)
```

```python
# We rely on the plain vanilla Stochastic Gradient Descent as our optimizing methodology
model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
```

```python
history1 = model.fit(X_train, y_train,
                     batch_size = batch_size,
                     epochs = training_epochs,
                     verbose = 2,
                     validation_data=(X_cv, y_cv))
```

```python
# Using adam optimizer
Inp = Input(shape=(784,))
x = Dense(n_hidden_1, activation='relu', name = "Hidden_Layer_1")(Inp)
x = Dense(n_hidden_2, activation='relu', name = "Hidden_Layer_2")(x)
x = Dense(n_hidden_3, activation='relu', name = "Hidden_Layer_3")(x)
x = Dense(n_hidden_4, activation='relu', name = "Hidden_Layer_4")(x)
output = Dense(num_digits, activation='softmax', name = "Output_Layer")(x)

# We rely on ADAM as our optimizing methodology
adam = keras.optimizers.Adam(learning_rate=learning_rate)
model2 = Model(Inp, output)

model2.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
history2 = model2.fit(X_train, y_train,
                      batch_size = batch_size,
                      epochs = training_epochs,
                      verbose = 2,
                      validation_data=(X_cv, y_cv))
```
