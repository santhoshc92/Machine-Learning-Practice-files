# IQR method

Now with the IQR method we will be removing the outlier


The interquartile range is calculated in much the same way as the range. All you do to find it is subtract the first quartile from the third quartile: IQR = Q3 â€“ Q1. The interquartile range shows how the data is spread about the median


(data >= Q1 - 1.5*IQR) & (data <= Q3 + 1.5*IQR)

Where 1.5 is standard range set to check the outliers

So with the help of above formula and IQR value we can remove the outliers ,to remove the outliers we have to calculate the values within upper and lower limit

IQR=Q3-Q1

lower limit=Q1-1.5*IQR
upper limit=Q3+1.5*IQR

so consider for age column

Q3 = data.age.quantile(0.75)
IQR = Q3 - Q1
print(Q1-1.5*IQR,Q3+1.5*IQR)

so we'll consider values between (9.5 , 69.5) and rest all values will be outliers.

```python
# âœ… Corrected Code (Best Practice)
# âœ” Detect outliers for ALL columns first
# âœ” Remove rows only ONCE
# âœ” No distribution shifts
# âœ” Clean and reliable

col_names = ['age', 'duration', 'campaign', 'cons.conf.idx']

# Step 1: Create a mask for outliers (False = keep, True = outlier)
outlier_mask = pd.Series(False, index=data.index)

for name in col_names:
    Q1 = data[name].quantile(.25)
    Q3 = data[name].quantile(.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Mark rows that are outliers for this column
    outlier_mask = outlier_mask | (data[name] < lower) | (data[name] > upper)

# Step 2: Remove outlier rows ONCE
data = data[~outlier_mask]
# Note:
# The union across columns is done incrementally during the loop, using | on a single Series. No separate DataFrame needed.
# This method is cleaner, avoids creating a temporary DataFrame, and reduces memory usage.
```

# 1.VIF

A variance inflation factor (VIF) is  a measure of multicollinearity among the independent variables in a multiple regression model.


Finding  multicollinearity is important because while multicollinearity does not reduce the explanatory power of the model, it does reduce the statistical significance of the independent variables.


A large variance inflation factor (VIF) on an independent variable indicates a highly collinear relationship to the other variables that should be considered or adjusted for in the structure of the model and selection of independent variables.


So we'll consider the VIF values <=5 ,rest all we'll remove the featuers with value >5. (at times consider upto 6.5)

*** drop columns with highest VIF first (removing this alone will affect other column VIF) and perform VIF again and drop columns with higher VIF again and so on until VIF until >5.
We can't drop all highere VIF at once because columns are interconnected


```python


from statsmodels.stats.outliers_influence import variance_inflation_factor
col_list = []
for col in data.columns:
    if ((data[col].dtype != 'object') & (col != 'y') ):    # dtype or dtypes
        col_list.append(col)    # need to pass col in append

X = data[col_list]
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
print(vif_data)
```

# need better understanding of the block

confusion matrix: confusion_matrix(y_test,y_pred) # follow this sequence   
classification_report(y_test,y_pred) # follow this sequence. if sequence changed, FP and FN interchange  

order also based on ascending (0,1) in classification report. 1 (yes) and 0(no) are provided by us based on dataset  
False -0 / True - 1 for actual values,  -->       0 -Negative,  1 - Positive for predicted values  

Output format in python  
              predicted  values  
                 0         1    
Actual 0        TN         FP   
values 1        FN         TP  

eg: TP - 'n' people have cancer and model also predicted the same  
    FN - model says person doesn't have cancer but in reality cancer is there  
    FP - model says person has cancer but in reality no cancer  
    TN - 'n' people who don't have cancer and model accurately predicted as no cancer  

specificity: proportion of true negatives present in data  
TNR - True negative rate or specificity = TN/(TN+FP)  
**
sensitivity or  TPR- True positive rate or recall = TP/(TP+FN)  ..out of all actual positive records, how many are predicted as positive
      
False positive rate : what portion of negative class got incorrectly classified. eg:healthy people tagged  as sick(covid)  
       FPR  = FP/(TN+FP) = 1 - specificity  

False negative rate : what portion of positive class got incorrectly classified. eg: sick classified as healthy  
       FNR  = FN/(TP+FN)  
precision  = TP/(TP+FP)  ..out of all records predicted as positive, how many are actually positive  
 
 Accuracy (all correct)= (TP+TN)/(TP+FP+TN+FN)  
 misclassification (all incorrect) = (FP+FN)/(TP+TN+FP+FN)  
 
 precision --> out of predicted record, how many are true  
 recall ---> out of actual record, how many predicted true  

 **** precision and recall are never use TN

# precision recall curve : to find optimal cutoff value

```python
import numpy as np
from sklearn.metrics import precision_recall_curve

# precision recall curve for test set
y_pred_proba = logisticRegr.predict_proba(x_test)[:,1]

# calculate precision and recall values
precision, recall, thresholds= precision_recall_curve(np.array(y_test), y_pred_proba)
f1_scores = 2 * (precision * recall) / (precision + recall + 1e-6)

best_cutoff = thresholds[np.argmax(f1_scores)]
print('Best_Cutoff:',best_cutoff )

# plot the precision and recall curve
plt.figure(figsize=(8,6))
plt.plot(thresholds, precision[1:],label = 'precision')
plt.plot(thresholds, recall[1:],label = 'recall')
plt.plot(thresholds, f1_scores[1:], label='F1 Score', linestyle='--') # optional not necessary 

# Add legends and label
plt.legend()
plt.xlabel('threshold')
plt.ylabel('score')
plt.title('Precision, Recall, F1 Score vs Threshold')
plt.grid(True)
plt.show()

# if we take for 100 % recall , precision is low
# Why we use precision[1:] and recall[1:]??
# The first precision/recall point (index 0) does NOT have a corresponding threshold. It represents the case where the threshold = 0 (model predicts everything as positive).
```
